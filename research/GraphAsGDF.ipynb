```python
import osmnx as ox
import geopandas
```

```python
def get_graph_as_gdf(place_name: str) -> tuple[geopandas.geodataframe.GeoDataFrame, geopandas.geodataframe.GeoDataFrame]:
    G = ox.graph_from_place(place_name, network_type="drive")
    return ox.utils_graph.graph_to_gdfs(G)
```

```python
nodes, edges = get_graph_as_gdf("Петергоф")
```
